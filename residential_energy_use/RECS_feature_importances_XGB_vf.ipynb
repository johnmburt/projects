#### Purpose of this notebook:

The Residential Energy Consumption Survey was conducted in 2017 by the U.S. Energy Information Administration (EIA). The survey asked an enormous number of questions about energy consumption, building features, appliances used, and economic and other demographic information about the home occupants. 

There are many facets to analyze and questions that could be asked of this dataset, but I chose to focus on a different topic than energy use: poverty. What factors in residential life, appliances, home construction and occupant demographics are associated with financial hardship? Knowing what factors predict poverty can help agencies identify people who may need extra help. Similarly, struggling people may be better helped by targeted upgrade assistance for obsolete appliances they're likely to own.

This notebook examines the relationship between residential and demograpic factors and economic hardship. I do this by training a classifier to predict whether a customer is experiencing economic hardship, and then looking at the most important features for that classification. 

Method: 
- Create a hardship score based on three features:
    - SCALEB, Frequency of reducing or forgoing basic necessities due to home energy bill
    - SCALEG, Frequency of keeping home at unhealthy temperature
    - SCALEE, Frequency of receiving disconnect notice
    - Use a score threshold to create a new binary variable "hardship"


- Create an input dataset using the other features, but dropping features:
    - Related to economics,
    - With string variables, NaNs or meta-features


- Train XGBoost classifier with selected features to predict economic hardship.


- Plot and examine the most important features.


- Discussion of results

### Load the data

```python
# remove warnings
import warnings
warnings.filterwarnings('ignore')
# ---

%matplotlib inline
import pandas as pd
pd.options.display.max_columns = 100
from matplotlib import pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import numpy as np

from datetime import datetime

pd.options.display.max_rows = 100

srcdir = './data/'

# load the data
df = pd.read_csv(srcdir+'recs2015_public_v4.csv') 

# load code keys
codes = pd.read_csv(srcdir+'codebook_publicv4.csv',
                    encoding = "ISO-8859-1")

# rename the useful columns
codes = codes.rename(columns = {
    'SAS Variable Name':'name', 
    'Type':'type', 
    'Variable Description':'description', 
    'Unnamed: 4':'values',
    'Final Response Set':'value_keys',
   })

# index on value name
codes = codes.set_index('name')
```

### Resample the data 

Each sample represents a specified number of households. I resample the data to create the same proportional frequencies.

```python
# Code provided by Kenny Liao
def resample_data(data):
    # calculate the number of times to duplicate each sample
    weights_scaled = ((data['NWEIGHT']/data['NWEIGHT'].min())).astype(int)
    # duplicate the original indices based on weights_scaled
    resampled_idx = data.index.repeat(weights_scaled.values)
    # create dummy dataframe with duplicated index and join original data
    resampled_data = pd.DataFrame(index=resampled_idx, columns=['dummy']).join(data)
    # delete dummy column and reset index
    resampled_data = resampled_data.drop('dummy', axis=1).reset_index(drop=True)
    return resampled_data

df = resample_data(df)

print('dataset size:',df.shape)
df.head()
```

### Remap column values for SCALE* features used for the hardship score

I need to swap the coding values so that the numbers describe an ordinal range, from 0 (no months per year) to 3 (every month per year). 

Currently:
- 1 = Almost every month 
- 2 = Some months
- 3 = 1 or 2 months
- 0 = Never

Re-map to:
- 0 = Never
- 1 = 1 or 2 months
- 2 = Some months
- 3 = Almost every month 


```python
# transform mapping
xform = {0:0, 1:3, 2:2, 3:1}

# remap the variables
df['SCALEBf'] = df['SCALEB'].map(xform)
df['SCALEGf'] = df['SCALEG'].map(xform)
df['SCALEEf'] = df['SCALEE'].map(xform)
```

### Histograms and correlations

To combine more than one feature using PCA, each must be significantly correlated with the others. I'll use a Spearman rank test to verify this requirement.

#### Result:
All of the variables are significantly positively correlated. 

The histograms also indicate that most samples are 0 (no economic distress), so this is going to be an imbalanced category problem.

```python
from scipy.stats import spearmanr

def print_corr(df, var1, var2):
    """Calc Spearman corr and print result"""
    coef, pval = spearmanr(df[var1],df[var2])
    print('%s vs %s: coef=%1.3f, pval=%1.5f'%(var1,var2,coef,pval))

print('Spearman rank correlations:\n')
print_corr(df,'SCALEBf','SCALEGf')
print_corr(df,'SCALEBf','SCALEEf')
print_corr(df,'SCALEGf','SCALEEf')

print('\nHistograms:')

df[['SCALEBf','SCALEGf','SCALEEf']].hist();
plt.tight_layout();
```

### Generate a PCA "hardship score"

Significant correlations between all vars indicates I can combine them to make a single score using PCA.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

# create PCA object
pca = PCA(n_components=1)

# do the PCA and save combined score to new 'hardship' column
scorefeatures = ['SCALEBf','SCALEGf','SCALEEf']
ycols = scorefeatures + ['SCALEB','SCALEG','SCALEE']

df['hardship'] = pca.fit_transform(df[scorefeatures])

# scale the hardship values between 0 - 1
df['hardship'] = MinMaxScaler().fit_transform(
    df['hardship'].values.reshape(-1, 1))

# plot histo of new hardship score
df['hardship'].hist();
plt.title('"hardship" - PCA combined score of SCALEB, SCALEG, SCALEE');

plt.figure()
df['hardship'].hist(range=[.2,1]);
plt.title('"hardship" - PCA combined score - only level >= 0.2');
```

## Select the input features

For this analysis I'm dropping any features related to economic status, since we already know they'll have high feature importance in any classifier model. 

I'll also drop metadata cols (start with Z or BRRW), cols with strings and Nulls.

For my target, I'm defining a binary variable based on thresholding the hardship score: 0 = below threshold (no hardship), 1 = above threshold (hardship).

```python
hardship_feat = [
    'SCALEBf','SCALEGf','SCALEEf',
    'SCALEB','SCALEG','SCALEE',
    'NOHEATDAYSf', 'NOHEATHELPf', 
    'NOACDAYSf', 'NOACHELPf',
    
    'COLDMA', 'HOTMA',
    'NOACBROKE', 'NOACEL',
    'NOHEATDAYS', 'NOHEATHELP',
    'NOACDAYS', 'NOACHELP',
    'NWEIGHT', 'DOEID',
    'NOHEATBROKE', 'NOHEATEL', 'NOHEATNG',
    'PAYHELP', 'FREEAUDIT', 'BENOTHER',
    'ENERGYASST', 'ENERGYASST11', 'ENERGYASST12', 
    'ENERGYASST13', 'ENERGYASST14', 'ENERGYASST15',
    'ENERGYASSTOTH', 'REBATEAPP', 'RECYCAPP', 'TAXCREDITAPP',
    'MONEYPY', 'EELIGHTS'
    ]

# drop features related to hardship
# drop columns that start with Z
# drop columns that start with BRRW
dropcols = ycols + hardship_feat + ['hardship']
Xcols = []
for s in df.columns:
    if s not in dropcols:
        if s[0] != 'Z' and 'BRRW' not in s:
            Xcols.append(s)
            
X = df[Xcols]
print(X.shape)            

# drop features with string values
okidx = []
for i in range(X.shape[1]):
    if type(X.iloc[0,i]) != str:
        okidx.append(i)
X = X[X.columns[okidx]]

# drop columns with NaNs
X = X.dropna(axis=1) 

print(X.shape)

# our target variable
# make it binary
threshold = 0.4
y = df['hardship'] > threshold
print(y.shape)

print('# samples: no hardship =', (~y).sum(), ', hardship =',y.sum())
print('Percent: no hardship =', 100*(~y).sum()/len(y), 
      ', hardship =',100*y.sum()/len(y))
```

## Using XGBoost to assess feature importances

Create a XGBoost model and train it with the data.

Cross-validate the model to assess accuracy.

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

# use scale_pos_weight parameter to account for class imbalance
spweight = (~y).sum() / y.sum()
# spweight = 1

# create model instance
clf = XGBClassifier(scale_pos_weight=spweight)

# get cross val score
scores = cross_val_score(clf, X, y, cv=3, scoring='balanced_accuracy')

print('Cross-validate model, metric = balanced accuracy:')
print('CV scores = ', ['%1.2f,'%(s) for s in scores])
print('mean score = %1.2f'%(np.mean(scores)))

# re-fit to entire dataset
clf.fit(X, y);
```

## Print and plot the most important features

```python
names = X.columns
numimportant = 20

importances = clf.feature_importances_
# std = np.std([tree.feature_importances_ for tree in clf.estimators_],
#              axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(numimportant):
    print("%d. (%1.3f) %s - %s" % (
        f + 1, 
        importances[indices[f]],
        names[indices[f]], 
        codes.loc[names[indices[f]]]['description'],
         ))

# Plot the feature importances of the forest
plt.figure(figsize=[10,5])
plt.title("Feature importances")
plt.bar(range(numimportant), importances[indices[:numimportant]],
       color="r", align="center")
plt.xticks(range(numimportant), names[indices[:numimportant]], rotation=90)
plt.xlim([-1, numimportant])
```

### Demographics

- 3. (0.022) KOWNRENT - Own or rent

- People having trouble paying bills tend to be less educated (the #1 predictor), are renters, female and younger. All of these demographic factors, except age, are well known correlates to poverty. The younger age factor is interesting: I would have predicted that older people on fixed incomes would experience more hardship. It's possible that those people are eligable to receive more assistance and take advantage of this (a future analysis topic).



### Building aspects
- 2. (0.022) DRAFTY - Frequency of draft
- 4. (0.018) OUTLET - Outlet within 20 feet of vehicle parking
- 8. (0.012) LGTOUTNUM - Number of light bulbs installed outside the home
- 10. (0.011) TYPEHUQ - Type of housing unit
- 12. (0.011) LGTINNUM - Number of light bulbs installed inside the home
- 13. (0.011) BASEHEAT - Heating used in basement
- 19. (0.010) BASECOOL - Cooling used in basement

- The first factor to consider here is that people in economic hardship are much more likely to live in mobile homes and appartments. These factors predict the other factors: they don't have basements or the kind of property that has a lot of lightbulbs outside. They live in smaller spaces with fewer rooms so have fewer lightbulbs indoors too. Their dwellinsg tend to be poorly insulated and drafty - signs of older, poorly maintained and run-down buildings that the poor are forced to live in. 



### Appliances
- 5. (0.017) ELPERIPH - Number of printers, scanners, fax machines, or copiers
- 6. (0.013) TVONWD1 - Most-used TV usage on weekdays
- 7. (0.012) TVTYPE2 - Display type of second most-used TV

- 17. (0.010) DOLELTV2 - Electricity cost for second televisions, in dollars, 2015

- Poor people tend not to have computers and their peripherals, or fax machines. Their TVs are older, using older tech like vacuum tubes, which use a lot more energy. They also watch a lot more TV, likely throughout the day, indicating lack of employment. I hadn't thought of this before, but TV watching habits are probably an excellent proxy indicator of employment status. 


### Climate
- 9. (0.011) GNDHDD65 - Heating degree days of ground temperature in 2015, base temperature 65F
- 15. (0.010) GWT - Annual average ground water temperature (F) for 2015


- Poor people are more likely to live in warmer areas. I attribute this to the demographic trend of higher poverty rates in Southern states. Unfortunately, Southern state governments also tend to provide less assistance of any kind.



## Comparing values of important features<br>hardship vs no hardship

### Notes: 
- I'm plotting % of each sub-sample (hardship vs no hardship) so that the proportions are comparable on the graphs.
- The further down the importance list, the less dissimilar the two categories are.
- Most of the values for each feature are codes, not numeric ranges, and some are inverted from what you'd expect!

```python
from textwrap import wrap
from matplotlib import gridspec

def get_value_key(values, keys):
    if values != values:
        return ''
    try:
        values = values.split('\n')
        keys = keys.split('\n')
        keystr = ('Value key:\n\n' + 
            '\n'.join(['%s - %s'%(s1,'\n'.join(wrap(s2,25))) for s1,s2 in zip(
                values,keys)]) )
    except:
        keystr = 'Value key:\n\n' + keys
    return keystr

imp_colnames = names[indices[:numimportant]]

for name in imp_colnames:
    titlestr = '\n'.join(wrap(codes.loc[name]['description'], 50))
    xrange = [X[name].min(), X[name].max()]
    plt.figure(figsize=[10,3])
    gs = gridspec.GridSpec(1, 2, width_ratios=[3, 1]) 
    ax = plt.subplot(gs[0])
    h1, bins1 = np.histogram(X[name][y==0])
    h2, bins2 = np.histogram(X[name][y==1], bins=bins1)
#     X[name][y==0].hist(label='No hardship', range=xrange)
    width = (bins1[1] - bins1[0])/3
    ax.bar(bins1[:-1], 100*h1/np.sum(h1), width=width, 
           facecolor='cornflowerblue', label='No hardship')
    ax.bar(bins2[:-1]+width, 100*h2/np.sum(h2),width=width, 
           facecolor='lightsalmon', label='Hardship')
    plt.ylabel('% of subsamples')
    plt.legend()
    plt.title(titlestr);
    
    plt.subplot(gs[1])
    plt.axis('off')
    plt.text(0,0.2,get_value_key(codes.loc[name]['values'],
                               codes.loc[name]['value_keys']))
```
